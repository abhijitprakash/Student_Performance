```python
# construct box plot for continuous variables
cont=df.dtypes[(df.dtypes!='uint8') & (df.dtypes!='bool')].index
plt.figure(figsize=(10,10))
df[cont].boxplot(vert=0)
plt.title('With Outliers',fontsize=16)
plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[cont].columns:
    lr,ur=remove_outlier(df[columns_num])
    df[columns_num]=np.where(df[columns_num]>ur,ur,df[columns_num])
    df[columns_num]=np.where(df[columns_num]<lr,lr,df[columns_num])
```

```python
plt.figure(figsize=(10,10))
df[cont].boxplot(vert=0)
plt.title('After Outlier Removal',fontsize=16)
plt.show()
```

```python
for feature in df.columns: 
    if df[feature].dtype == 'object': #categorical to numerical
        print('\n')
        print('feature:',feature)
        print(pd.Categorical(df[feature].unique()))
        print(pd.Categorical(df[feature].unique()).codes)
        df[feature] = pd.Categorical(df[feature]).codes
```

```python
X = df.drop(["G3"], axis=1)

Y= df["G3"]
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
plt.scatter(y_train_pred,  y_train_pred - Y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred,  y_test_pred - Y_test, c='red', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=0, xmax=20, lw=2, color='cyan')
plt.xlim([0, 20])
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
print('MSE train: %.3f, test: %.3f' % (
        mse(Y_train, y_train_pred),
        mse(Y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2(Y_train, y_train_pred),
        r2(Y_test, y_test_pred)))
print('mae train:%.3f,test: %.3f' %(
       mae(Y_train, y_train_pred),
       mae(Y_test, y_test_pred)))
```
